```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
```

```python
plt.rcParams["figure.figsize"] = (15, 10)
```

```python
def random_walk(p, n, k):
    x = np.random.binomial(1, p, (k, n)) * 2 - 1
    return np.cumsum(x, axis=1);
```

```python
s = random_walk(0.5, 1000, 10)
for row in s:
    plt.plot(row)
plt.plot(np.mean(s, axis=0), color="black", linewidth=2);
```

```python
p = 0.5
```

```python
s = random_walk(p, 1000, 100)
m = s * s - np.arange(1000)
for row in m:
    plt.plot(row)
plt.plot(np.mean(m, axis=0), color="black", linewidth=2);
```
